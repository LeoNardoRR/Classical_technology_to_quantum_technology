```python
import matplotlib.pyplot as plt

# Dados para o gráfico
x = [1, 2, 3]  # Valores do eixo x (por exemplo, pontos no tempo em segundos)
y1 = [0.001, 0.037, 303.24]  # Dados da primeira série (segundos)
y2 = [0, 0, 0]  # Dados da segunda série (segundos)
y3 = [0, 0, 0]  # Dados da terceira série (segundos)
y4 = [0.000211, 0.000120, 0.000113]  # Dados da quarta série (segundos)

# Criar o gráfico de linhas comparativo
plt.plot(x, y1, label='Java - Fermat')
plt.plot(x, y2, label='Java - Shor')
plt.plot(x, y3, label='Q# - Fermat')
plt.plot(x, y4, label='Q# - Shor')

# Aplicar escala logarítmica no eixo y para séries com valores pequenos
plt.yscale('log')

# Adicionar rótulos e título
plt.xlabel('Tempo (segundos)')
plt.ylabel('Escala')
plt.title('Gráfico Comparativo')

# Adicionar uma legenda
plt.legend()

# Exibir o gráfico
plt.show()
```
